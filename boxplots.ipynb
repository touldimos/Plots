```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def color_boxes(df):
    vals, names, xs = [],[],[]
    for i, col in enumerate(df.columns):
        vals.append(df[col].values)
        names.append(col)
        xs.append(np.random.normal(i + 1, 0.04, df[col].values.shape[0]))

    plt.boxplot(vals, labels=names)
    # palette = ['b', 'g', 'y', 'r', 'purple', 'darkorange', 'turquoise']
    palette = ['navy', 'darkcyan', 'darkviolet', 'crimson', 'darkorange', 'gold', 'forestgreen']

    for x, val, c in zip(xs, vals, palette):
        plt.scatter(x, val, alpha=0.4, color=c)
#     plt.ylim(0.5, 1)
    plt.xticks(rotation = 45);
```

```python
df = pd.DataFrame(np.random.rand(4, 100).T, columns = ['a', 'b', 'c', 'd'])
color_boxes(df)
```
