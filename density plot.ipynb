```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import pandas as pd

data = pd.read_csv(r"C:\Users\owner\Google Drive 2\MSc\TU DELFT - Water Managment\Additional Thesis\Results\45.txt", delimiter = '\t')
# Generate fake data
x = data['x']
y = data['y']

# Calculate the point density
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sort the points by density, so that the densest points are plotted last
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]

fig, ax = plt.subplots(figsize=(19.20, 10.80))
ax.scatter(x, y, c = z, s = 50, zorder = 2)
ax.set_xscale('log')
ax.set_xlim(0.1, x.max())
ax.tick_params(labelsize = 15)
ax.set_ylabel((u'$E_{r}$ (mm\xb7 month $^{-1}$)'), fontsize  = 15)
ax.set_xlabel((u'$c_{th}$ (-)'), fontsize  = 15)
plt.grid(zorder = 1)
plt.show()
fig.savefig(r'C:\Users\owner\Google Drive 2\MSc\TU DELFT - Water Managment\Additional Thesis\Results\45.png', format='png', dpi=1200)
```
