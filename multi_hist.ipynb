```python
import pandas as pd
import numpy as np

df = pd.read_excel('book1.xlsx', index_col = 0).T
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
A = np.array(df.loc['A'].drop('Count'))
B = np.array(df.loc['B'].drop('Count'))
C = np.array(df.loc['C'].drop('Count'))
D = np.array(df.loc['D'].drop('Count'))
E = np.array(df.loc['E'].drop('Count'))
H = np.array(df.loc['H'].drop('Count'))
All = np.array(df.loc['-'].drop('Count'))
```

```python
plotdata = pd.DataFrame({
    "A":A,
    "B":B,
    "C":C,
    "D":D,
    "E":E,
    "H":H,
    "All":All,
    }, 
    index=["Mean", "St. Error (*100)", "Median", "S. Deviation", "S. Variance", "Range", "Minimum", "Maximum"]
)
plotdata.plot(kind="bar", figsize = (15, 5), zorder = 2)
plt.title("Correction coefficient per Koppen Class")
plt.xlabel("Value")
plt.ylabel("Statistics")
plt.grid(zorder = 1);
```
